```python
#전부 출력
def brute_force_matrix_sum(matrix):
    w = len(matrix[0])
    h = len(matrix)
    maxSum = 0
    start_w = 0
    start_h = 0
    end_w=0
    end_h=0
    for j in range(h):
        for i in range(w):
            for l in range(h-j):
                for k in range(w-i):
                    temp=0
                    for n in range(l+1):
                        for m in range(k+1):
                            temp += matrix[j+n,i+m]
                    if temp >maxSum:
                        maxSum = temp
                        start_w=i
                        start_h=j
                        end_w=i+k
                        end_h=j+l
    print("start : ({},{}) , end : ({},{}), max :{}".format(start_w,start_h,end_w,end_h,maxSum))
                    
                
```

```python
#전부 출력
def divide_conquer_matrix_sum(matrix):
    h = len(matrix)
    w = len(matrix[0])    
    lu=(0,(0))
    ru=(0,(0))
    ld=(0,(0))
    rd=(0,(0))
    start_h = 0
    start_w = 0
    end_h = 0
    end_w = 0
    
    if w==1 and h==1:
        maxSum = matrix[0,0]
        start_h = 0
        start_w = 0
        end_h = 0
        end_w = 0
        return (maxSum,(start_h,start_w,end_h,end_w))
    
    if h==w:
        lu=divide_conquer_matrix_sum(matrix[:-1,:-1])
        ru=right_up(matrix[:-1,:],0)
        ld=left_down(matrix[:,:-1],0)
        rd=right_down(matrix)
    else:
        if h<w:
            lu=divide_conquer_matrix_sum(matrix[:,:-1])
            ru=right_up(matrix,0)
        else:
            lu=divide_conquer_matrix_sum(matrix[:-1,:])
            ld=left_down(matrix,0)
            
    maxSum = max(lu[0],ru[0],ld[0],rd[0])
    
    if lu[0] == max(lu[0],ru[0],ld[0],rd[0]):
        return lu
    elif ru[0] == max(lu[0],ru[0],ld[0],rd[0]):
        return ru
    elif ld[0] == max(lu[0],ru[0],ld[0],rd[0]):
        return ld
    else:
        return rd

def right_up(matrix,index):
    h = len(matrix)
    w=len(matrix[0])
    #mid point
    if h==1:
        temp = 0
        tempMax =-999
        for i in range(w):
            temp += matrix[0,w-i-1]
            if tempMax < temp:
                tempMax = temp
                start_h = 0+index
                start_w = w-i-1
                end_h = 0+index
                end_w = w-1
        return (tempMax,(start_h,start_w,end_h,end_w))
    
    mid = int(h/2)
    
    #cross
    #크로스맥스일 경우 포함하는 거 구하기
    crossMax = 0
    for i in range(mid+1):
        for j in range(h-mid):
            temp = 0
            for k in range(w):
                temp += np.sum(matrix[mid-i:mid+j+1,w-k-1])
                if crossMax<temp:
                    crossMax = temp
                    start_h = mid-i
                    start_w = w-k-1
                    end_h = mid+j
                    end_w = w-1
    
    crossMax=(crossMax,(start_h,start_w,end_h,end_w))
    
    #up
    upMax = right_up(matrix[:mid,:],0)
    #down
    if mid != h-1:
        downMax = right_up(matrix[mid+1:,:],mid+1+index)
    else:
        downMax = (0,(0))
        
    if crossMax[0] == max(crossMax[0],upMax[0],downMax[0]):
        return crossMax
    elif upMax[0] == max(crossMax[0],upMax[0],downMax[0]):
        return upMax
    else:
        return downMax

def left_down(matrix,index):
    w = len(matrix[0])
    h=len(matrix)
    if w==1:
        temp = 0
        tempMax = -999
        for i in range(h):
            temp+= matrix[h-i-1,0]
            if tempMax < temp:
                tempMax = temp
                start_h = h-i-1
                start_w = 0+index
                end_h = h-1
                end_w = 0+index
        return (tempMax,(start_h,start_w,end_h,end_w))
    mid = int(w/2)
    #cross
    crossMax = 0
    for i in range(mid+1):
        for j in range(w-mid):
            temp = 0
            for k in range(h):
                temp += np.sum(matrix[h-k-1,mid-i:mid+j+1])
                if crossMax<temp:
                    crossMax = temp
                    start_h = h-k-1
                    start_w = mid-i
                    end_h = h-1
                    end_w = mid+j
    crossMax=(crossMax,(start_h,start_w,end_h,end_w))
                    
    #left
    leftMax = left_down(matrix[:,:mid],0)
    #right
    if mid != w-1:
        rightMax = left_down(matrix[:,mid+1:],mid+1+index)
    else:
        rightMax = (0,(0))
        
    if crossMax[0] == max(crossMax[0],leftMax[0],rightMax[0]):
        return crossMax
    elif leftMax[0] == max(crossMax[0],leftMax[0],rightMax[0]):
        return leftMax
    else:
        return rightMax


def right_down(matrix):
    h=len(matrix)
    w=len(matrix[0])
    downMax = 0
    memory = np.zeros((h,1))
    temp = np.zeros((h,1))
    for i in range(w-1,-1,-1):
        for j in range(h-1,-1,-1):
            temp[j,0] = matrix[j,i]
            if j!=h-1:
                temp[j,0] = matrix[j,i]+temp[j+1,0]
            if temp[j,0]+memory[j,0]>downMax:
                downMax = temp[j,0]+memory[j,0]
                start_h = j
                start_w = i
                end_h = h-1
                end_w = w-1
        memory = memory+temp
    return (downMax,(start_h,start_w,end_h,end_w))
```
